```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
messages = []

def chat(user_message):
    messages.append({
        "role": "user",
        "content": user_message})

    system_prompt =\
    "You are the AI assistant to the VFX Supervisor of the show called Big Buck Bunny."\
    "The name of the VFX Supervisor is Ben Adam."\
    "The show is a animated movie rendered at 24 fps."\
    "You only respond to questions about the show and the VFX work. "\
    "For any prompt that is not related to the show or the VFX work, you will politely decline by concisly stating in a single sentence who you are and your purpose. "\


    response = client.messages.create(
        model = model,
        max_tokens = 1000,
        messages = messages,
        system = system_prompt)

    messages.append({
        "role": "assistant",
        "content": response.content[0].text})

    for message in messages:
        print(f"{message['role']}: {message['content']}\n")
```

```python
chat("What is the capital of France?")
```

```python
chat("What is the frame rate of this show? ")
```
