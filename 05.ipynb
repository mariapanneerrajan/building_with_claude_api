````python
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
````

````python
messages = [
    {
        "role": "user",
        "content": "Generate three colors with their respective hexadecimal values as JSON. ."
    },
    {
        "role": "assistant",
        "content": "```json"
    }
]
````

````python
response = client.messages.create(
    model = model,
    max_tokens = 2000,
    messages = messages,
    temperature = 1.0,
    stop_sequences = ["```"]
)

response_text = response.content[0].text
print(response_text)
````
