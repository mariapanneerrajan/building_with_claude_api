```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
messages = []

def chat(user_message):
    messages.append({
        "role": "user",
        "content": user_message})

    response = \
        client.messages.create(
        model = model,
        max_tokens = 1000,
        messages = messages,
        temperature = 1.0,)

    messages.append({
        "role": "assistant",
        "content": response.content[0].text})

    for message in messages:
        print(
        f"{message['role']}:"\
        f"{message['content']}\n")
```

```python
chat("Generate a one-sentence movie idea")
```
